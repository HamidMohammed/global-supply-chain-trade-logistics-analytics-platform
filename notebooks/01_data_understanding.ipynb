```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
shipping_df = pd.read_csv("/Users/hamid/Desktop/SupplyChain_Project/data/shipping_rates.csv")
port_df = pd.read_csv("/Users/hamid/Desktop/SupplyChain_Project/data/port_congestion.csv")
trade_df = pd.read_csv("/Users/hamid/Desktop/SupplyChain_Project/data/trade_flows.csv")
events_df = pd.read_csv("/Users/hamid/Desktop/SupplyChain_Project/data/disruption_events.csv")
industry_df = pd.read_csv("/Users/hamid/Desktop/SupplyChain_Project/data/industry_exposure.csv")
```

```python
def audit_dataframe(df, name):
    print(f"\n{'='*60}")
    print(f"DATASET: {name}")
    print(f"{'='*60}")
    
    print("\nShape:")
    print(df.shape)
    
    print("\nColumns:")
    print(df.columns.tolist())
    
    print("\nData Types:")
    print(df.dtypes)
    
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())
    
    print("\nSample Data:")
    display(df.head())
```

```python
audit_dataframe(shipping_df, "Shipping Rates")
audit_dataframe(port_df, "Port Congestion")
audit_dataframe(trade_df, "Trade Flows")
audit_dataframe(events_df, "Disruption Events")
audit_dataframe(industry_df, "Industry Exposure")
```
